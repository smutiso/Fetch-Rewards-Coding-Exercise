## First: Review unstructured JSON data and diagram a new structured relational data model

Using pandas I will converted these unstructured JSON files into cleaned structured data.

```python
# Import Required Libraries

import pandas as pd
from ast import literal_eval
import json
from datetime import datetime
from sqlalchemy import create_engine
```

```python
# Read the json files

receipts = pd.read_json('receipts.json',lines=True)
brands = pd.read_json('brands.json',lines=True)
users = pd.read_json('users.json',lines=True)
```

### Receipts table.

```python
receipts.head()
```

The rewardsReceiptItemList field contains lists of receipts therefore we use the literal_eval function from the ast package to explode it. Then using the json_normalize function in pandas we split each dictionary and get the columns from it. Finally, we join this data back to the original receipts data frame.   

```python
receipts = receipts.explode('rewardsReceiptItemList')
receipts.reset_index(inplace=True)
```

```python
receipts = receipts.fillna({'rewardsReceiptItemList':'{}'})
receipts['rewardsReceiptItemList'] = receipts['rewardsReceiptItemList'].apply(lambda x:str(x))
receipts['rewardsReceiptItemList'] = receipts['rewardsReceiptItemList'].apply(literal_eval)
#raises an exception if the input isn't a valid Python datatype, so the code won't be executed if it's not.
```

```python
receipts_norm = pd.json_normalize(receipts['rewardsReceiptItemList'],errors='ignore',record_prefix='rewardsReceiptItemList')\
.add_prefix('rewardsReceiptItemList.')
```

```python
#Normalize semi-structured JSON data into a flat table.
final_receipts = pd.merge(receipts, receipts_norm, left_index=True, right_index=True, how='outer')
```

we observe that each of these columns contain only one dictionary item each so we need to get rid of all the unwanted characters. As for the date columns the time is stored as UTC (Coordinated Universal Time) so we convert that into datetime format using the apply function.

```python
def date_converter(x):
  try:
    return(datetime.utcfromtimestamp(int(x['$date'])/1000).strftime('%Y-%m-%d %H:%M:%S'))
  except TypeError:
    return(None)
```

```python
final_receipts['_id'] = final_receipts['_id'].apply(lambda x: x['$oid'])
final_receipts['createDate'] = final_receipts['createDate'].apply(lambda x: date_converter(x))
final_receipts['dateScanned'] = final_receipts['dateScanned'].apply(lambda x: date_converter(x))
final_receipts['finishedDate'] = final_receipts['finishedDate'].apply(lambda x: date_converter(x))
final_receipts['modifyDate'] = final_receipts['modifyDate'].apply(lambda x: date_converter(x))
final_receipts['pointsAwardedDate'] = final_receipts['pointsAwardedDate'].apply(lambda x: date_converter(x))
final_receipts['purchaseDate'] = final_receipts['purchaseDate'].apply(lambda x: date_converter(x))
```

Example of a query: Which brand has the most spend among users who were created within the past 6 months?

```python
engine.execute("SELECT * from (SELECT r.rewardsReceiptItemList_brandCode,SUM(r.totalSpent) as summedTotalSpent from receipts r where r.userId in (SELECT _id from users u where u.createdDate > (SELECT strftime('%Y-%m-%d %H:%M:%S', 'now','-6 month'))) group by r.rewardsReceiptItemList_brandCode) where rewardsReceiptItemList_brandCode is not NULL order by summedTotalSpent DESC LIMIT 1;").fetchall()
```
